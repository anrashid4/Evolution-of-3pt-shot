```python
# File to load 
file_to_load = './Resources/nba_shot_type.csv'
```

```python
nba_shots = pd.read_csv(file_to_load)
```

```python
nba_shots
```

```python
nba_shots = nba_shots[['YEAR', 'PLAYER', 'PCT_FGA_2PT', 'PCT_FGA_3PT', 'MIN']]
```

```python
nba_shots
```

```python
# Only keeping 3 seasons
nba_shots1 = nba_shots[nba_shots['YEAR'] == '2016-2017']
```

```python
# Only keeping 3 seasons
nba_shots2 = nba_shots[nba_shots['YEAR'] == '2017-2018']
```

```python
# Only keeping 3 seasons
nba_shots3 = nba_shots[nba_shots['YEAR'] == '2018-2019']
```

```python
nba_shots1
```

```python
nba_shot_type = pd.concat([nba_shots1, nba_shots2, nba_shots3], axis=0)
```

```python
nba_shot_type
```

```python
nba_shot_type['YEAR'] = nba_shot_type['YEAR'].replace(['2016-2017','2017-2018','2018-2019'],['2016-17','2017-18','2018-19'])
```

```python
nba_shot_type
```

```python
nba_data
```

```python
mergedDf = nba_shot_type.merge(nba_data)
```

```python
mergedDf
```

```python
mergedDf.set_index('YEAR')
```

```python
mergedDf = mergedDf.set_index('YEAR')
```
